```python
from bs4 import BeautifulSoup
url = 'https://www.footballdb.com/teams/index.html'
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)
soup = BeautifulSoup(r.text, 'html.parser')

print(soup.prettify())
```

```python
# Get NFL team list
division_list = [
    'AFC East',
    'AFC North',
    'AFC South',
    'AFC West',
    'NFC East',
    'NFC North',
    'NFC South',
    'NFC West']
team_list = list()
rows = soup.find_all('tr', {'class': re.compile('row*')})
for r in rows:
    tds = r.find_all('td')
    for td in tds:
        d = dict()
        if td.find('b'):
            d['team_name'] = td.b.a.text
            url = td.b.a['href']
            d['url'] = url
            for l in ['roster', 'stats', 'results', 'draft', 'transaction', 'history']:
                d[l] = f"{url}/{l}"
        if d and 'nfl' in url:
            team_list.append(d)
with open('team_list.json', 'w') as fp:
    json.dump(team_list, fp, indent=4)
```
